```python
import numpy as np

# Example array
arr = np.array([3, 1, 2, 5, 4])

# Get indices that would sort the array in ascending order
sorted_indices = np.argsort(arr)
print("Indices for ascending order sorting:", sorted_indices)

# Sort the array using the indices
sorted_arr = arr[sorted_indices]
print("Array sorted in ascending order:", sorted_arr)

# Get indices that would sort the array in descending order
sorted_indices_desc = np.argsort(arr)[::-1]
print("Indices for descending order sorting:", sorted_indices_desc)

# Sort the array in descending order using the indices
sorted_arr_desc = arr[sorted_indices_desc]
print("Array sorted in descending order:", sorted_arr_desc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 15, 20, 25],
    'Value2': [5, 8, 12, 18]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create bar plot using Seaborn
sns.set_style("whitegrid")
ax = sns.barplot(x='Category', y='Value1', data=df, color='blue')

# Create secondary y-axis
ax2 = ax.twinx()
sns.barplot(x='Category', y='Value2', data=df, ax=ax2, color='red')

# Set labels and title
ax.set_ylabel('Value1', color='blue')
ax2.set_ylabel('Value2', color='red')
plt.title('Bar Plot with Two Y-Axes')

plt.show()
```
